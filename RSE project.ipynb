```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('area_used in sq km.csv')
df.head()
```

```python
df.info()
```

```python
df.replace(to_replace='-',value='0',inplace=True)
```

```python
pd.options.display.max_columns = 90
pd.options.display.max_rows = 9999
```

```python
df['Baden-Württemberg (Area in sq km)'] = df['Baden-Württemberg (Area in sq km)'].astype(float)
df['Bayern (Area in sq km)'] = df['Bayern (Area in sq km)'].astype(float)
df['Berlin (Area in sq km)'] = df['Berlin (Area in sq km)'].astype(float) 
df['Brandenburg (Area in sq km)'] = df['Brandenburg (Area in sq km)'].astype(float) 
df['Bremen (Area in sq km)'] = df['Bremen (Area in sq km)'].astype(float)
df['Hamburg (Area in sq km)'] = df['Hamburg (Area in sq km)'].astype(float) 
df['Hessen (Area in sq km)'] = df['Hessen (Area in sq km)'].astype(float) 
df['Mecklenburg-Vorpommern (Area in sq km)'] = df['Mecklenburg-Vorpommern (Area in sq km)'].astype(float) 
df['Niedersachsen (Area in sq km)'] = df['Niedersachsen (Area in sq km)'].astype(float) 
df['Nordrhein-Westfalen (Area in sq km)'] = df['Nordrhein-Westfalen (Area in sq km)'].astype(float) 
df['Rheinland-Pfalz (Area in sq km)'] = df['Rheinland-Pfalz (Area in sq km)'].astype(float) 
df['Saarland (Area in sq km)'] = df['Saarland (Area in sq km)'].astype(float) 
df['Sachsen (Area in sq km)'] = df['Sachsen (Area in sq km)'].astype(float) 
df['Sachsen-Anhalt (Area in sq km)'] = df['Sachsen-Anhalt (Area in sq km)'].astype(float) 
df['Schleswig-Holstein (Area in sq km)'] = df['Schleswig-Holstein (Area in sq km)'].astype(float) 
df['Thüringen (Area in sq km)'] = df['Thüringen (Area in sq km)'].astype(float)
```

```python
df.info()
```

# State which has consumed most area in total:

```python
df2 = df.T
df2.head()
```

```python
print('len: ',len(df2.index))
print(df2.iloc[1])
print(df2.iloc[1].sum())
```

```python
sum_of_area_of_each_state = []

for i in range(len(df2.index)-1):
    if i == 0:
        pass
    else:
        sum_of_area_of_each_state.append(df2.iloc[i].sum())

print('array for sum of area of each state: ',sum_of_area_of_each_state)
print('length of array: ',len(sum_of_area_of_each_state))      
```

```python
sum_for_each_state = {}
# print(df.columns)
j = 0

for i in df.columns:
    if i == "Types of use" or i == "Year":
        pass
    else:
        sum_for_each_state[i] = sum_of_area_of_each_state[j]
        j += 1

print("sum_for_each_state", sum_for_each_state)        
```

```python
# Create figure and axis objects
plt.figure(figsize=(14, 10))

# Plotting the data
plt.bar(sum_for_each_state.keys(), sum_for_each_state.values(), color='skyblue')

# Adding title and labels
plt.title('Total Area covered for Each State in Germany')
plt.xlabel('State')
plt.ylabel('Total Area (sq km)')

# Rotating x-axis labels for better readability
plt.xticks(rotation=60, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

# Purpose which has consumed most area in total

```python
df.head()
```

```python
df3 = df.groupby('Types of use').sum()
df3
```

```python
df3.iloc[0].sum()
```

```python
sum_of_area_of_each_purpose = []

for i in range(len(df3.index)):
    sum_of_area_of_each_purpose.append(df3.iloc[i].sum())

print('array for sum of area of each purpose: ',sum_of_area_of_each_purpose)
print('length of array: ',len(sum_of_area_of_each_purpose))      
```

```python
df3.index
```

```python
sum_for_each_purpose = {}
# print(df.columns)
j = 0

for i in df3.index:
    sum_for_each_purpose[i] = sum_of_area_of_each_purpose[j]
    j += 1

print("sum_for_each_purpose: ", sum_for_each_purpose)        
```

```python
# Create figure and axis objects
plt.figure(figsize=(14, 10))

# Plotting the data
plt.bar(sum_for_each_purpose.keys(), sum_for_each_purpose.values(), color='orange')

# Adding title and labels
plt.title('Total Area covered for Each purpose in Germany')
plt.xlabel('Purpose')
plt.ylabel('Total Area (sq km)')

# Rotating x-axis labels for better readability
plt.xticks(rotation=60, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
